from our datasets we have unecessary column 

```python
Sentiment_data = Sentiment_data.drop(columns=["Unnamed: 0.1", "Unnamed: 0"])
Sentiment_data.head()
```

then we need to fix the timestamp column

```python
Sentiment_data["Timestamp"] = pd.to_datetime(Sentiment_data["Timestamp"])
```

we fit to date and time columns

```python
Sentiment_data["Date"] = Sentiment_data["Timestamp"].dt.date
Sentiment_data.head()
```

```python
Sentiment_data["Time"] = Sentiment_data["Timestamp"].dt.time
Sentiment_data.head()
```

then we delect the timestamp column 

```python
Sentiment_data = Sentiment_data.drop(columns=["Timestamp"])
Sentiment_data.head()
```

then our data is cleaned we save the cleaned data

```python
Sentiment_data.to_csv("Cleaned_Sentiment_data.csv", index=False)
```
